### Notebook de Carga de Datos
Este notebook se encarga de cargar el archivo Excel extraidos mediante el scraping y renombrar las columnas. Los resultados se guardan en un archivo procesado dentro de data/processed/.

```python
import pandas as pd

# Cargar el dataset
datos_excel = pd.read_excel('../data/raw/boletin_2016_01-2024_08.xlsx')

# Renombrar columnas manualmente
new_column_names = [
    'Año', 'Mes', 'Día', 'Temp Media', 'Temp Máxima', 'Hora Temp Max', 
    'Minuto Temp Max', 'Temp Mínima', 'Hora Temp Min', 'Minuto Temp Min', 
    'Viento Medio', 'Dirección Viento', 'Ráfaga de Viento', 'Dirección de Ráfaga', 
    'Hora Viento', 'Minuto Viento', 'Presión Media', 'Lluvia Diaria', 
    'Radiación Máxima', 'Hora Radiación Máxima', 'Minuto Radiación Mínima', 
    'UV Máximo', 'Hora UV Máximo', 'Minuto UV Máximo'
]

# Asignar los nuevos nombres
datos_excel.columns = new_column_names

# Guardar el archivo procesado
datos_excel.to_csv('../data/processed/datos_procesados.csv', index=False)

# Mostrar los primeros registros
datos_excel.head()
```
