```python
import pandas as pd

# Cargar el dataset
datos_excel = pd.read_excel('../data/raw/boletin_2016_01-2024_08.xlsx')
datos_excel.head()
```

```python
# Ver los nombres de las columnas
datos_excel.columns

# Renombrar columnas manualmente
new_column_names = [
    'Año',  # Variable A
    'Mes',  # Variable M
    'Día',  # Variable D
    'Temp Media',  # temp media
    'Temp Máxima',  # temp max
    'Hora Temp Max',  # H tmax
    'Minuto Temp Max',  # M tmax
    'Temp Mínima',  # temp min
    'Hora Temp Min',  # H tmin
    'Minuto Temp Min',  # M tmin
    'Viento Medio',  # Viento en km/h
    'Dirección Viento',  # dir
    'Ráfaga de Viento',  # rfga
    'Dirección de Ráfaga',  # dir (de la ráfaga)
    'Hora Viento',  # H (Hora del viento)
    'Minuto Viento',  # M (Minuto del viento)
    'Presión Media',  # pres. media
    'Lluvia Diaria',  # lluvia diaria
    'Radiación Máxima',  # rad max
    'Hora Radiación Máxima',  # H rad
    'Minuto Radiación Mínima',  # M rad
    'UV Máximo',  # UV max
    'Hora UV Máximo',  # H UVmax
    'Minuto UV Máximo'  # M UVmax
]

# Asignar los nuevos nombres (asegúrate de tener el mismo número de nombres)
datos_excel.columns = new_column_names

# Mostrar el DataFrame con los nuevos nombres de columna
print(datos_excel.head())
```
